# Zomato Dataset Exploratry Data Annalysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Importing the dataset in pandas dataframe
df = pd.read_csv('zomato.csv',encoding='latin-1')
```

```python
# Checking for null values in the dataset
df.isnull().sum()
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
#adding new feature Country to the dataframe
df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
df.head()
```

```python
df.dtypes
```

## Q.1 Which Country Has the most number of restaurants listed on Zomato?

```python
country_restaurant_count = df['Country'].value_counts()
most_restaurants_country = country_restaurant_count.idxmax()
print(most_restaurants_country, "has the most number of restaurants listed on Zomato")
```

```python
# # Plotting a bar chart to visualize the restaurant count by country
plt.figure(figsize=(12, 6))
sns.countplot(x='Country', data=df, order=df['Country'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Restaurant Count')
plt.title('Number of Restaurants by Country')

# Add percentage count labels to the bars
bars = plt.bar(country_restaurant_count.index, country_restaurant_count.values)
total_restaurants = len(df)
for bar in bars:
    height = bar.get_height()
    percentage = (height / total_restaurants) * 100
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{percentage:.1f}%', ha='center', va='bottom')
```

## Q.2 Which city has the most number of restaurants listed on Zomato in india?

```python
# Filter the dataset for restaurants in India
df_india = df[df['Country'] == 'India']

# Count the number of restaurants by city
city_restaurant_count = df_india['City'].value_counts()

# Get the city with the most number of restaurants
most_restaurants_city = city_restaurant_count.idxmax()

print("City with the most number of restaurants in India:", most_restaurants_city)
```

```python
df_india.head()
```

```python
most_restaurants_cities = city_restaurant_count.head(5)
plt.figure(figsize=(8, 8))
plt.pie(most_restaurants_cities, labels=most_restaurants_cities.index, autopct='%1.1f%%')
plt.title('Top 5 cities has the most number of restaurants listed on Zomato in india')
plt.axis('equal')
```

## Q.3 What are the most common types of cuisines available on Zomato?

```python
cuisine_counts = df['Cuisines'].str.split(', ').explode().value_counts()
```

```python
top_cuisines = cuisine_counts.head(5)
print("Most common cuisines:", top_cuisines)
```

```python
plt.figure(figsize=(8, 8))
plt.pie(top_cuisines, labels=top_cuisines.index, autopct='%1.1f%%')
plt.title('Top 5 Most Common Cuisines on Zomato')
plt.axis('equal')
```

## Q.4 What are the top-rated restaurants in each city?

```python
top_rated_restaurants = df.groupby('City').apply(lambda x: x.nlargest(1, 'Aggregate rating'))
print("Top-rated restaurants in each city:\n", top_rated_restaurants[['City', 'Restaurant Name', 'Aggregate rating']])
```

```python
top_rated_restaurants.head()
```

```python
# Plotting a map to visualize the top-rated restaurants by city
import folium
```

```python
map_top_rated_restaurants = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=12)

for index, row in top_rated_restaurants.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['Restaurant Name'],
        icon=folium.Icon(color='green')
    ).add_to(map_top_rated_restaurants)

map_top_rated_restaurants.save('top_rated_restaurants_map.html')
```
